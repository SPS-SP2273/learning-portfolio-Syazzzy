# Task 1

```python
dic = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

filename = 'oct-08_singapore.txt'

filename = filename.replace(filename[:3], dic[filename[:3]])

filename
```

# Task 2

```python
def rename_my_file(old_file_name):
    return old_file_name.replace(old_file_name[:3], dic[old_file_name[:3]])

print(rename_my_file('oct-08_singapore.txt'))
print(rename_my_file('apr-10_manila.txt'))
print(rename_my_file('aug-05_shanghai.txt'))
```

# Task 3

```python
old_file_names = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

new_file_names = []

for file in old_file_names:
    new_file_name = rename_my_file(file)
    new_file_names.append(new_file_name)
    print(file + ' ----> ' + new_file_name)
    
```

# Task 4

```python
import os
import glob
import shutil

txt_files = glob.glob('os-collaboration-exercise-data/**/*.txt', recursive=True)

for path in txt_files:
    old_file_name = path.split(os.path.sep)[-1]
    new_file_name = path.split(os.path.sep)[0] + '\\' + rename_my_file(old_file_name)
    os.rename(path, new_file_name)

glob.glob('os-collaboration-exercise-data/**/*.txt', recursive=True)
```

# Task 5

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    path = os.path.join('os-collaboration-exercise-data', city)
    print(f'Creating {path}')
    os.mkdir(path)
```

# Task 6

```python
cairo_files = glob.glob('os-collaboration-exercise-data/**/*cairo*', recursive=True)

for path in cairo_files:
        
    current_path = os.path.join(path)
    new_path = os.path.join(path.split(os.path.sep)[0], 'Cairo')

    shutil.move(current_path, new_path)
    print(f'Moved {current_path} to {new_path}')
```

# Task 7

```python
for city in cities[1:]:
    city_files = glob.glob(f'os-collaboration-exercise-data/**/*{city}*', recursive=True)
    for path in city_files:
        current_path = os.path.join(path)
        new_path = os.path.join(path.split(os.path.sep)[0], city)

        shutil.move(current_path, new_path)
        print(f'Moved {current_path} to {new_path}')
```
